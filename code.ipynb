<h1>Load Dataset</h1>

```python
import pandas as pd
import numpy as np

# Generating 1000 synthetic rows
np.random.seed(42)

df = pd.DataFrame({
    "Altitude": np.random.randint(28000, 40001, 1000),
    "Wind_Speed": np.random.randint(50, 160, 1000),
    "Pressure": np.random.randint(260, 340, 1000),
    "Temp_Gradient":np.round(np.random.uniform(2.0, 10.0, 1000), 2),
    "Wind_Shear": np.round(np.random.uniform(1.0, 12.0, 1000), 2),
    "Vertical_Acceleration": np.round(np.random.uniform(0.01, 0.25, 1000), 3),
    "Jet_Stream_Speed": np.random.randint(110, 190, 1000)
})

#Turbulence label logic (FIXED)
df["Turbulence"] = (
    (df["Wind_Shear"] > 8) |
    (df["Jet_Stream_Speed"] > 140) |
    (df["Temp_Gradient"] > 7)
).astype(int)

#Saved as CSV File
path = "/Users/aaseaansiwakoti/AI/Coursework/turbulence_dataset.csv"
df.to_csv(path, index=False)

path
```
